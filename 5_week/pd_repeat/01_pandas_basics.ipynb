## 1. Introduction to Pandas

```python
import numpy as np
import pandas as pd

# Creating a Series with explicit index labels
obj = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"], dtype='float32')
print("Original Series:\n", obj)

# Reordering based on index labels
print("\nSeries reordered by index labels:\n", obj[["c", "a", "d"]])

# Scalar multiplication - returns a new Series, original remains unchanged
print("\nSeries multiplied by 2 (original unchanged):\n", obj * 2)
```

## 2. Introduction to Pandas DataFrame

```python
import pandas as pd
import numpy as np

# Creating a DataFrame from a dictionary
data = {"state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
"year": [2000, 2001, 2002, 2001, 2002, 2003],
"pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data)
print("DataFrame created from dictionary:\n", frame)

# Creating a DataFrame with specified columns, adding a new column 'debt'
frame = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'])
frame['debt'] = np.arange(6)
print("\nDataFrame with new 'debt' column:\n", frame)

# Assigning a Series to a DataFrame column, handling missing values with NaN
frame['debt'] = pd.Series([1.2, 2, 3.3], index=[0, 2, 4])
print("\nDataFrame after assigning a Series to 'debt' (NaN for missing indices):\n", frame)
```

## 3. Pandas Index Operations

```python
import pandas as pd

# Append two Index objects
idx1 = pd.Index([1, 2, 3])
idx2 = pd.Index([4, 5, 6])
idx3 = idx1.append(idx2)
print("Appended Index:\n", idx3)

# Difference between two Index objects
idx4 = pd.Index([1, 2, 3, 4, 5])
idx5 = pd.Index([4, 5])
idx6 = idx4.difference(idx5)
print("\nDifference (elements in idx4 but not in idx5):\n", idx6)

# Intersection of two Index objects
idx7 = pd.Index([1, 2, 3, 4])
idx8 = pd.Index([3, 4, 5, 6])
idx9 = idx7.intersection(idx8)
print("\nIntersection (common elements):\n", idx9)

# Delete element from an Index by position
idx10 = pd.Index([1, 2, 3, 4, 5])
idx11 = idx10.delete(2)  # Delete element at position 2 (value 3)
print("\nIndex after deleting element at position 2:\n", idx11)
```

## 4. Reindexing with `ffill` (Forward Fill)

```python
import pandas as pd
import numpy as np

obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
print("Original Series for reindexing:\n", obj3)

# Reindex to a new index range and forward-fill missing values
obj3 = obj3.reindex(index=np.arange(6), method="ffill")
print("\nSeries after reindex with ffill:\n", obj3)
```

## 5. Dropping Rows and Columns from DataFrame

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
print("Original DataFrame:\n", data)

# Drop rows (returns a copy, original DataFrame 'data' remains unchanged)
print("\nDataFrame after dropping 'Colorado' and 'Ohio' rows:\n", data.drop(index=["Colorado", "Ohio"]))

# Drop columns (returns a copy, original DataFrame 'data' remains unchanged)
print("\nDataFrame after dropping 'one' and 'three' columns:\n", data.drop(columns=["one", "three"]))
```

## 6. Selection with `loc` (label-based) and `iloc` (position-based)

```python
import pandas as pd
import numpy as np

obj = pd.Series(np.arange(4., 8.), index=["a", "b", "c", "d"])
print("Original Series:\n", obj)

# Select elements by index labels in a specific order
print("\nSeries selection using .loc with labels:\n", obj.loc[["b", "a", "d"]])

# Select elements by integer position
print("\nSeries selection using .iloc with positions:\n", obj.iloc[[0, 1, 2]])
```

## 7. DataFrame Indexing and Slicing

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
print("Original DataFrame:\n", data)

# Filter rows based on a column's value
print("\nRows where 'two' > 6:\n", data[data['two'] > 6])

# Select the first two rows
print("\nFirst two rows:\n", data[:2])

# Boolean indexing to select elements less than 2 (returns DataFrame with NaN for False values)
print("\nElements < 2:\n", data[data < 2])

# Select specific row by label and specific columns by label
a = data.loc['Colorado', ['two', 'three']]

# Select specific row by position and specific columns by position
b = data.iloc[2, [3, 0, 1]]

# Slice rows up to 'Utah' (inclusive) and select 'two' column
c = data.loc[:"Utah", "two"]

print("\nSelection 'a' (Colorado, 'two', 'three'):\n", a)
print("\nSelection 'b' (iloc[2], [3, 0, 1]):\n", b)
print("\nSelection 'c' (loc[:'Utah', 'two']):\n", c)
```

## 8. DataFrame Arithmetic and Alignment

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns=list("abcd"))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)), columns=list("abcde"))

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)

# Addition with alignment based on row and column labels (NaN for non-matching)
print("\nDataFrame 1 + DataFrame 2 (with alignment):\n", df1 + df2)

# Addition with fill_value for non-matching labels
print("\nDataFrame 1.add(DataFrame 2, fill_value=0):\n", df1.add(df2, fill_value=0))
```

## 9. Applying Functions to DataFrame (Column-wise and Row-wise)

```python
import pandas as pd
import numpy as np

# Custom function to calculate range (max - min)
def f1(x):
    return x.max() - x.min()

data = pd.DataFrame(np.random.randn(5, 4), columns=list("ABCD"))
print("Original DataFrame:\n", data)

# Apply function column-wise (default axis=0)
for_each_column = data.apply(f1)
print("\nResult of applying f1 to each column:\n", for_each_column)

# Apply function row-wise (axis=1 or axis='columns')
for_each_line = data.apply(f1, axis=1)
print("\nResult of applying f1 to each row:\n", for_each_line)

# Custom function returning a Series (results in new DataFrame rows)
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])

print("\nResult of applying f2 to each column (returns Series):\n", data.apply(f2))
```

## 10. Element-wise Operation with `map`

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.randn(5, 4), columns=list("ABCD"))
print("Original DataFrame for map:\n", data)

# Custom function for element-wise transformation
def f3(x):
    if x > 0:
        return 100
    else:
        return -100

# Apply f3 element-wise to the DataFrame
print("\nDataFrame after element-wise mapping with f3:\n", data.map(f3))
```

## 11. Sorting DataFrame Values

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.randn(5, 4), columns=list("ABCD"))
print("Original DataFrame for sorting:\n", data)

# Sort by a single column 'D'
aa = data.sort_values('D')
print("\nDataFrame sorted by column 'D':\n", aa)

# Sort by multiple columns 'B' then 'C'
bb = data.sort_values(['B', 'C'])
print("\nDataFrame sorted by columns 'B' and 'C':\n", bb)
```

## 12. Ranking DataFrame Values

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.randn(5, 4), columns=list("ABCD"))
print("Original DataFrame for ranking:\n", data)

# Rank values in each column (default ascending)
ranked = data.rank()
print("\nDataFrame with ranks (default ascending):\n", ranked)

# Rank values in descending order
ranked = data.rank(ascending=False)
print("\nDataFrame with ranks (descending):\n", ranked)
```

## 13. Time Series Data Simulation and Basic Analysis

```python
import pandas as pd
import numpy as np

np.random.seed(0)

# Create a date index and tickers for stock prices
dates = pd.date_range('2016-10-13', '2016-10-21')
tickers = ["AAPL", "GOOG", "IBM", "MSFT"]

# Simulate a simple random-walk price series for each ticker
start_prices = np.array([115.0, 770.0, 135.0, 55.0])
daily_returns = np.random.normal(loc=0.0, scale=0.01, size=(len(dates), len(tickers)))
prices = (1 + daily_returns).cumprod(axis=0) * start_prices

price = pd.DataFrame(prices, index=dates, columns=tickers)
print("Simulated Stock Prices (first 5 rows):\n", price.head())
```

## 14. Calculating Returns, Correlation, and Covariance

```python
import pandas as pd
import numpy as np

# (Assuming 'price' DataFrame from previous cell is available)

# Calculate percent change (daily returns)
returns = price.pct_change()
print("Daily Returns (last 5 rows):\n", returns.head())

# Correlation and covariance between MSFT and IBM returns
msft_ibm_corr = returns["MSFT"].corr(returns["IBM"])
msft_ibm_cov = returns["MSFT"].cov(returns["IBM"])
print(f"\nMSFT-IBM correlation: {msft_ibm_corr}")
print(f"MSFT-IBM covariance: {msft_ibm_cov}")

# Full correlation matrix of all stock returns
print("\nCorrelation Matrix:\n", returns.corr())

# Full covariance matrix of all stock returns
print("\nCovariance Matrix:\n", returns.cov())
```
